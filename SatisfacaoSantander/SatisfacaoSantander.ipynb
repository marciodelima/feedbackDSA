```python
# Importando as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

%matplotlib inline
warnings.filterwarnings("ignore")
```

```python
# Importando o arquivo 
df = pd.read_csv('data/train.csv')
```

```python
df.head(10)
```

```python
df.dtypes
```

```python
#Dados estatisticos - analise descritiva
df.describe()
```

```python
#Tratamento dos dados NA - Missing
df.fillna(0)
```

```python
#Dados de Correlação entre as variáveis
df.corr()
```

```python
#Construindo um gráfico de HEATMAP
f, ax = plt.subplots(figsize=(15, 12))
sns.heatmap(df.corr(),linewidths=.5, ax=ax)
```

```python
# Distribuição da varíavel TARGET
df.groupby('TARGET').size()
```

```python
df.shape
```

```python
#Limpando a coluna ID
df = df.drop("ID", axis=1)
df.shape
```

```python
# Separando os dados
array = df.values

# Separando o array em componentes de input (X) e output (Y)
X = array[:,0:369]
Y = array[:,369]

df_new = df.drop("TARGET", axis=1)
```

```python
#Feature Selection - Escolhendo as 20 melhores variáveis / fatores utilizando CHI2
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import SelectPercentile, f_classif

best_var = SelectKBest(f_classif, k=20)
fitColunas = best_var.fit(X, Y)
# Pegando as colunas escolhidas
cols = fitColunas.get_support()
print(df_new.columns[cols])

df_selecao = df[ df_new.columns[cols] ]
X_new = fitColunas.transform(X)
X_new.shape
```

```python
#Gráficos
df_selecao.hist()
plt.show()
```

```python
df_selecao.plot(kind = 'density', subplots = True, layout = (8,3), sharex = False)
plt.show()
```

```python
df_selecao.plot(kind = 'box', subplots = True, layout = (8,3), sharex = False, sharey = False)
plt.show()
```

```python
# Normalizando e padronizando os dados do dataframe

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

# Gerando a nova escala (normalizando os dados)
scaler = MinMaxScaler(feature_range = (0, 1))
rescaledX = scaler.fit_transform(X_new)

#Gerando padronização dos valores
scalerP = StandardScaler().fit(rescaledX)
standardX = scalerP.transform(rescaledX)
```

```python
# Cross-Validation
from sklearn.model_selection import KFold

# Definindo os valores para os folds
num_folds = 10
seed = 1313

# Separando os dados em folds
kfold = KFold(num_folds, True, random_state = seed)
```

```python
#Criando o Modelo - Versao 1
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

modelo = LogisticRegression()
```

```python
#Resultado do Modelo - Versao 1 - Utilizando o Cross Validation
resultado = cross_val_score(modelo, standardX, Y, cv = kfold)

# Print do resultado
print("Acuracia do Modelo 1: %.3f" % (resultado.mean() * 100))
```

```python
#Criando o Modelo - Versao 2 - XGBClassifier e gerando dados de Treino e de Teste
#!pip install xgboost
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split

#Gerando os dados de Treino e Teste
X_treino, X_teste, y_treino, y_teste = train_test_split(standardX, Y, test_size = 0.35, random_state = seed)

#Criando o modelo
modelo = XGBClassifier()
```

```python
# Treinando o modelo
modelo.fit(X_treino, y_treino, eval_metric="auc", verbose = False)
```

```python
# Fazendo previsões
y_pred = modelo.predict(X_teste)
previsoes = [round(value) for value in y_pred]
```

```python
#Resultado do Modelo - Versao 2
resultado2 = accuracy_score(y_teste, previsoes)
print("Acuracia do Modelo 2: %.3f" % (resultado2 * 100.0))
```
